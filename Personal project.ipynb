### sales according to year

```python
Bike_sales['Year'].value_counts()
```

```python
Bike_sales['Year'].value_counts().plot(kind='line', figsize=(16,4))
plt.legend()
```

### sales according to month

```python
Bike_sales['Month'].value_counts()
```

```python
Bike_sales['Month'].value_counts().plot(kind='line', figsize=(12,6))
plt.legend()
```

# sales according to age group

```python
Bike_sales['Age_Group'].value_counts()
```

```python
Bike_sales['Age_Group'].value_counts().plot(kind= 'pie', figsize=(12,6), autopct='%1.1f%%', fontsize=12, shadow=True, explode=(0,0.1,0,0))
plt.title('sales according to age group')
plt.show()
```

# sales by product category

```python
Bike_sales['Product_Category'].value_counts()
```

```python
Bike_sales['Product_Category'].value_counts().plot(kind='bar', figsize=(12,6))
plt.xlabel('Product_Category')
plt.ylabel('count')
plt.title('Sales by Product Category')
plt.legend()
```

# sales by sub_category

```python

Bike_sales['Sub_Category'].value_counts()
```

```python
Bike_sales['Sub_Category'].value_counts().plot(kind='bar', figsize=(12,6))
plt.xlabel('Sub_category')
plt.ylabel('Sales')
plt.title('Sales by sub_categoery')
plt.legend
```
